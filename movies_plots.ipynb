```python
# Supress Warnings
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
movies=pd.read_csv('Movie+Assignment+Data.csv')
movies
```

```python
# Write your code for dropping the columns here. It is advised to keep inspecting the dataframe after each set of operations
movies=movies.drop(['color','director_facebook_likes','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes','actor_2_name','cast_total_facebook_likes','actor_3_name','duration','facenumber_in_poster','content_rating','country','movie_imdb_link','aspect_ratio','plot_keywords'],axis=1)
movies
```

```python
# Write your code for dropping the rows here
movies=movies[movies.gross.notnull()]
movies=movies[movies.budget.notnull()]
movies
```

```python
# Write your code for dropping the rows here
movies=movies[movies.isnull().sum(axis=1).sort_values(ascending=False) <=5]
movies
```

```python
movies.language=movies.language.fillna('English')
movies
```

```python
# Write your code for creating the profit column here
movies['profit']=movies['gross']-movies['budget']
# Write your code for dropping duplicate values here
movies=movies.drop_duplicates(keep='first')
movies
```

```python
# Write your code for sorting the dataframe here
top10=movies.sort_values('profit',ascending=False).head(10)
top10
```

```python
plt.xlabel('Profit (in millions)',fontsize=20,fontname='Times New Roman',color='#0d1af3',weight='bold')
plt.ylabel('Movie',fontsize=20,fontname='Times New Roman',color='pink',weight='bold')
plt.title('Top 10 Profit making movies',fontsize=20,fontname='Times New Roman',color='r',weight='bold')
plt.barh(top10['movie_title'],top10['profit'])
plt.show()
```

```python
top10director=movies.groupby('director_name')[['imdb_score']].mean().sort_values(by='imdb_score',ascending = False).head(10)
top10director
```

```python
plt.xlim(8,9)
plt.xlabel('Mean IMDB score',fontsize=20,fontname='Times New Roman',color='#0d1af3',weight='bold')
plt.ylabel('Director',fontsize=20,fontname='Times New Roman',color='k',weight='bold')
plt.title('Top 10 Directors',fontsize=20,fontname='Times New Roman',color='r',weight='bold')
plt.barh(top10director.index,top10director['imdb_score'])
plt.show()
```

```python
GenTemp=movies.genres.str.split('|',expand=True).iloc[:,0:2]
GenTemp.columns=['genre_1','genre_2']
GenTemp.genre_2.fillna(GenTemp.genre_1,inplace=True)
PopGenre=pd.concat([movies,GenTemp],axis=1)
p=PopGenre.groupby(['genre_1','genre_2'])[['gross']].mean().sort_values(by='gross',ascending=False).head(5)
l=list(p.index)
df=pd.DataFrame(l)
df
```

```python
plt.xlabel('Mean profit',fontsize=20,fontname='Times New Roman',color='#0d1af3',weight='bold')
plt.ylabel('Genre',fontsize=20,fontname='Times New Roman',color='k',weight='bold')
plt.title('Top 5 genres',fontsize=20,fontname='Times New Roman',color='r',weight='bold')
plt.barh(df[0]+"+"+df[1],p['gross'])
plt.show()
```

```python
# Write your code for creating three new dataframes here
Meryl_Streep=movies[movies['actor_1_name']=='Meryl Streep']
Leo_Caprio=movies[movies['actor_1_name']=='Leonardo DiCaprio']
Brad_Pitt=movies[movies['actor_1_name']=='Brad Pitt']
Combined=Meryl_Streep.append([Leo_Caprio,Brad_Pitt])
c=Combined.groupby('actor_1_name')['num_critic_for_reviews','num_user_for_reviews'].mean().sort_values(by='num_critic_for_reviews',ascending=False)
c
```

```python
plt.xlabel('Actor',fontsize=20,fontname='Times New Roman',color='#0d1af3',weight='bold')
plt.ylabel('Mean no. of critics for reviews',fontsize=16,fontname='Times New Roman',color='k',weight='bold')
plt.bar(c.index,c['num_critic_for_reviews'])
```

```python
plt.xlabel('Actor',fontsize=20,fontname='Times New Roman',color='#0d1af3',weight='bold')
plt.ylabel('Mean no. of users for reviews',fontsize=16,fontname='Times New Roman',color='k',weight='bold')
plt.bar(c.index,c['num_user_for_reviews'])
```
