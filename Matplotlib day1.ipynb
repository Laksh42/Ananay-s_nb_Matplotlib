```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,6,8,10]
y=[10,15,35,66,90]
plt.plot(x,y)
plt.show()
```

```python
x=[2,4,6,8,10]
y=[10,15,35,66,90]
plt.bar(x,y)
plt.show()
```

```python
x=[2,4,6,8,10]
y=[10,15,35,66,90]
plt.scatter(x,y)
plt.show()
```

```python
x=[2,4,6,8,10]
y=[10,15,35,66,90]
plt.barh(x,y)
plt.show()
```

```python
x=[2,4,6,8,10]
y=[10,15,35,66,90]
z=[15,34,76,97,34]

plt.title('First Plot')
plt.xlabel('Time')
plt.ylabel('Sensor')

plt.plot(x,y,label='Sensor1')
plt.plot(x,z,label='Sensor2')

plt.legend()
plt.show()
```

```python
x=[2,4,6,8,10]
y=[10,15,35,66,90]
z=[15,34,76,97,34]

plt.title('First Plot')
plt.xlabel('Time')
plt.ylabel('Sensor')

plt.bar(x,y,label='Sensor1')
plt.bar(x,z,label='Sensor2')

plt.legend()
plt.show()
```

```python
x=[2,4,6,8,10]
y=[10,15,35,66,90]
z=[15,34,76,97,34]

plt.title('First Plot')
plt.xlabel('Time')
plt.ylabel('Sensor')

plt.scatter(x,y,label='Sensor1')
plt.scatter(x,z,label='Sensor2')
plt.plot(x,y)
plt.plot(x,z)

plt.legend()
plt.show()
```
